# Concat Dataframes

```python
df_bangladesh=pd.DataFrame({
    'city':['dhaka','cumilla', 'chandpur', 'borishal'],
    'temperature':[23,34,45,34],
    'Humadity':[60,45,34,78]
    })
```

```python
df_bangladesh
```

```python
df_germany=pd.DataFrame({
    'city':['köln','Aachen', 'Düsseldorf', 'Essen'],
    'temperature':[12,11,45,34],
    'Humadity':[60,65,84,78]
    })
```

```python
df_germany
```

```python
df=pd.concat([df_germany,df_bangladesh])
#index are not following the sequences
```

```python
df
```

```python
#to keep the sequences countinue use ignor_index
df=pd.concat([df_germany,df_bangladesh],ignore_index=True)
```

```python
df
```

```python
df=pd.concat([df_germany,df_bangladesh],keys=['germany','Bangladesh'])
```

```python
df
```

```python
df=pd.concat([df_germany,df_bangladesh],axis=1)
```

```python
df
```
