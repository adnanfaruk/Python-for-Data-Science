```python
pv=fluege.pivot_table(values='passengers',index='month',columns='year')
```

```python
pv
```

```python
sns.heatmap(pv,cmap='coolwarm',linecolor='white',linewidths=1)
```

```python
sns.clustermap(pv,cmap='coolwarm',standard_scale=1)
```
